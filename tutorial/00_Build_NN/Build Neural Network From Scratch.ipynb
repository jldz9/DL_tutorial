In this dataset, each row of data is a hand written digit, each digit is 20x20 pixel large.The 20x20 pixel image have 400 pixels, which also means 400 features (400 rows).

Each picture was reshaped into one row in the training dataset. Let's try to plot them.

```python
def plot_single_image(one_row):
    image = one_row.reshape(20, 20).T
    plt.imshow(image, cmap='gray')
    plt.show()
one_row = x_train[1500,:] # This represent a single row of data in dataset
plot_single_image(one_row)
```
